```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
import imp

iris = datasets.load_iris()
plt.plot(iris.data[0:50, 0], iris.data[0:50, 1])
plt.show()
plt.scatter(iris.data[0:50, 0], iris.data[0:50, 1])
plt.show()
```

```python
#regresi memprediksi suatu atribut berdasarkan atribut yang lain
iris.data[0:10,0:2]
```

```python
X = np.c_[iris.data[:, 0], np.ones([50, 1])]
```

```python
X1 = iris.data[0:50, 0]
X1
```

```python
mones = np.ones(50)
mones
```

```python
X = np.c_[X1, mones]
X.shape
```

```python
y=iris.data[0:50, 1]
y
```

```python
r1 = np.matmul(X.T, y) #.T=transpose
r2 = np.matmul(X.T, X)
```

```python
r2inv = np.linalg.inv(r2)
b=np.matmul(r2inv, r1)
b
```

```python
XX = np.arange(0, 10, 0.02)
YY = b[0]*XX + b[1]
plt.scatter(XX, YY)
plt.show()
plt.scatter(iris.data[0:50, 0], iris.data[0:50, 1])
plt.show()
```

```python
r=iris.data[0:10, [0,1,2]]
r
```

```python
E1 = iris.data[0:50, [0,2]]
E = np.c_[E1, mones]
g1 = np.matmul(X.T, r) #.T=transpose
g2 = np.matmul(X.T, E)
g2inv = np.linalg.inv(g2)
h=np.matmul(g2inv, g1)
h
```
