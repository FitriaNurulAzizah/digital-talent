# 1. Missing Value
Sering kali, data rusak, atau hilang, kita perlu mengurusnya terlebih dahulu karena kedepannya data ini tidak berfungsi saat data hilang atau tidak lengkap.

# Imputing missing values dengan Imputer

```python
import pandas as pd
from sklearn.preprocessing import Imputer #mengambil dari library sklearn sub modulnya prepocessing untuk menghandle missing value
```

```python
df = pd.read_csv('Data.csv') #baca data
df #NaN adalah missing valuenya bisa jadi gak ada datanya, bisa jadi kosong (0) 
```

```python
df.isnull().sum() #melihat summary nya, null itu yang tidak ada datanya, dilihat ada 1 di age, ada 1 di salary yang perlu ditangani
```

```python
df.dropna() #data yang missing value dianggap sampah, jadi dihilangkan tidak disimpan cuma didrop
```

```python
# drop kolom spesifik yang mengandung NaN #cara lain selain di atas
df.dropna(subset=['Age'])
```

```python
df.dropna(subset=['Salary']) #untuk drop spesifik kolomnya misalnya salary
```

```python
df.iloc[:, 1:3] #indeks pertama menslice barisnya, indeks ke dua menslice kolomnya, mengambil semua baris hanya di kolom 1-3 (3tidak dihitung)
```

```python
# replace every occurrence of missing_values to one defined by strategy
# which can be mean, median, mode. Axis = 0 means rows, 1 means column
#imputer fungsi/class di sklearn yang digunakan untuk mereplace missing value tadi, misalnya diisi dengan rata-rata/median/mode

imputer = Imputer(missing_values='NaN', strategy='mean', axis = 0) #axis 0 artinya per kolom
df.iloc[:, 1:3] = imputer.fit_transform(df.iloc[:, 1:3]) #ditampung di dalam variabel imputer yang nantinya akan menggunakan fungsi fit_transform, 
df #panda dataframe diinputkan dengan fit transform

#salary germain di baris no4 diganti mean salary, age spain di baris no 6 diganti mean agenya
#axis = sumbu x
```

# 4. Fitur Scaling
normalisasi

```python

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler

df = pd.read_csv('Data.csv').dropna()
print(df)
X = df[["Age", "Salary"]].values.astype(np.float64)
print(X)

#memproses data yang numerik saja yaitu age dan salary
#NORMALISASI YANG X SAJA
```

```python
standard_scaler = StandardScaler()
normalizer = Normalizer()
min_max_scaler = MinMaxScaler()

print("Standardization")
print(standard_scaler.fit_transform(X))

print("Normalizing")
print(normalizer.fit_transform(X))

print("MinMax Scaling")
print(min_max_scaler.fit_transform(X))

#1 YANG PALING TINGGI
#0 YANG PALING RENDAH
#e-01 = 10^-1
#normalisasi data numerik jika ingin di range tertentu
```

# 5.1 Count Vectorizer

* vectorizer= merubah text menjadi vektor
* Corpora/korpus = kumpulan2 dokumen
* kmren bikin wordcount cuma 1 dokumen

```python
from sklearn.feature_extraction.text import CountVectorizer


docs = ["Mayur Mayur is a nice boy.", "Mayur rock! wohooo!", "My name is Mayur, and I am a Pythonista!"]
cv = CountVectorizer()
X = cv.fit_transform(docs) #inputnya list of docs
print(X)
print(cv.vocabulary_)
print(X.todense())
```

```python
# 2itu id katanya, a dihiraukan, mayur = 4, yang jauh ini munculnya berapa kali, todense itu sudah dalam bentuk matriks,
#pengkodean/indeksnya diurutkan berdasarkan kata di kalimatnya
#untuk ukuran matriksny disesuaikan dengan kalimat ug paling banyak, 
#jika kalimat pertama tidak mengandung 1 kata di klimat terpanjang maka diberi angka 0
```

# Dict Vectorizer
DictVectorizer melakukan mapping dari dictionry wordcount ke Vektor, memappingkan jika sudah punya dictionary

```python
from sklearn.feature_extraction import DictVectorizer

docs = [{"Aku": 1, "suka": 1, "makan": 2}, {"Aku": 1, "tidak": 1, "suka": 2, "makan": 3, "kambing": 1, "bakar": 2, "madu": 3}]
dv = DictVectorizer()
X = dv.fit_transform(docs)
print(X)
print(X.todense())
```

```python
#dictionary jadi vektor, kalimat di atas aku muncul beberapa ....
#0 itu indeks artinya kalimat pertama, 1 itu indeks artinya kalimat ke 2
```

# TfIdf Vectorizer:
Word Count (Term Frekuensi dikali dengan Inverse Dokumen Frekuensi),

Tutorial dapat dilihat pada link berikut: https://datascience.mipa.ugm.ac.id/id/representasi-teks-dalam-vektor-part-1/ https://datascience.mipa.ugm.ac.id/id/representasi-teks-dalam-vektor-part-2/

misalnya kata aku dan menang muncul sering di korpus bola, nanti nilainya besar sekali tpi tidak penting maknanya, yang penting itu misal kata Barcelona sring muncul di satu dokumen tpi tidak di dokumen lainya. kata yang selalu muncul di setiap dokumen (contoh nya tim) itu menjadi tidak penting sehingga menurunkan scor dari kata tim.

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

tfidf_vectorizer = TfidfVectorizer()
cv_vectorizer = CountVectorizer()
docs = ["Mayur is a Guitarist Guitarist", "Mayur is Cooker", "Mayur is Musician", "Mayur is also a programmer"]
X_idf = tfidf_vectorizer.fit_transform(docs)
X_cv = cv_vectorizer.fit_transform(docs)
print(X_idf.todense())
print(tfidf_vectorizer.vocabulary_)
print(X_cv.todense())
```
