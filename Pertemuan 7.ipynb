```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# Join ada 4 Jenis:

# 1. INNER JOIN

```python
run("""
SELECT 
    productCode, 
    productName,
    textDescription
FROM
    products AS t1
        INNER JOIN
    productlines AS t2 ON t1.productline = t2.productline
""")

#kolom yg menyamakan keduanya kolom productline, sehingga join diperhatikan productLine 
```

```python
run("""
SELECT 
    productCode, 
    productName,
    t1.productLine,
    textDescription
FROM
    products AS t1
        INNER JOIN
    productlines AS t2 ON t1.productline = t2.productline
""")

#kolom yg menyamakan keduanya kolom productline, sehingga join diperhatikan productLine 
#productLinenya dari t1
#yg keluar yg ada irisannya
```

# USING

```python
run("""
SELECT 
    productCode, 
    productName, 
    textDescription
FROM
    products
        INNER JOIN
    productlines USING (productline)
""")

#productLine dari t1 dan dari t2 disingkat menggunakan using
#tabel yg dijoin dengan nama kolom yg beda tidak bisa pakai using
```

```python
#join antara orders dengan orderdetails menggunakan inner join
run("""
SELECT orderNumber,quantityordered,priceEach,productcode
FROM
    orders
        INNER JOIN
    orderdetails USING (orderNumber)
""")
```

# Kombinasi dengan GROUP BY

```python
run("""
SELECT 
    T1.orderNumber,
    status,
    SUM(quantityOrdered * priceEach) total
FROM
    orders AS T1
        INNER JOIN
    orderdetails AS T2 ON T1.orderNumber = T2.orderNumber
GROUP BY orderNumber
""")

#order ini total belanjanya segini, order ini total belanjanya segini
```

# Join dengan Operasi Perbandingan Lain

```python
run("""
SELECT 
    orderNumber, 
    productName, 
    msrp, 
    priceEach
FROM
    products p
        INNER JOIN
    orderdetails o ON p.productcode = o.productcode
        AND p.msrp > o.priceEach
WHERE
    p.productcode = 'S10_1678'
""")

#msrp=harga jual yg dicanangkan oleh perusahaan
#priceEach=realitasnya
#melihat produk khusus yg s10_1678 yg harga jualnya lebih rendah dari harga seharusnya
#on bisa untuk operasi perbandingan lain
```

# 2. LEFT JOIN

```python
run("""
SELECT
 c.customerNumber,
 c.customerName,
 orderNumber,
 o.status
FROM
 customers c
LEFT JOIN orders o ON c.customerNumber = o.customerNumber
""")

#customer yg left join dengan order brrti yg customer yg tidak order tetap tampil
```

```python

run("""
SELECT 
    c.customerNumber, 
    c.customerName, 
    orderNumber, 
    o.status
FROM
    customers c
        LEFT JOIN
    orders o ON c.customerNumber = o.customerNumber
WHERE
    orderNumber IS NULL
""")
```

# WHERE vs ON

```python
run("""
SELECT 
    o.orderNumber, 
    customerNumber, 
    productCode
FROM
    orders o
        LEFT JOIN
    orderDetails USING (orderNumber)
WHERE
    orderNumber = 10123
""")

#yg atas dipisahin dulu baru dibatasi
#where dilakukan setelah adanya join
#on dilakukan sebelum adanya where
```

```python

run("""
SELECT 
    o.orderNumber, 
    customerNumber, 
    productCode
FROM
    orders o
        LEFT JOIN
    orderDetails d ON o.orderNumber = d.orderNumber
        AND o.orderNumber = 10123
""")

#yg dikiri semua ditampilkan apapun on nya
#yg gak ada relasi ditandakan dengan none
#meskipun yg bukan 10123 muncul tetapi tidak ada relasi di kanan
```

# 3. RIGHT JOIN

```python

run("""
SELECT 
    concat(e.firstName,' ', e.lastName) salesman, 
    e.jobTitle, 
    customerName
FROM
    employees e
        RIGHT JOIN
    customers c ON e.employeeNumber = c.salesRepEmployeeNumber
        AND e.jobTitle = 'Sales Rep'
ORDER BY customerName
""")

#semua customernya ditampilkan tetapi kan ada yg tidak berelasi sebagai sales rep, mungkin bisa dngan VP / manager
```

# 5. Self Join
dalam employee pasti punya 1 atasan, dan ada banyak bawahan, letaknya ada di satu tabel

```python
run("""
SELECT 
    CONCAT(m.lastname, ', ', m.firstname) AS 'Manager',
    CONCAT(e.lastname, ', ', e.firstname) AS 'Direct report'
FROM
    employees e
        INNER JOIN
    employees m ON m.employeeNumber = e.reportsto
ORDER BY manager
""")

#e alias dr employee pertama
#managernya bondur gerard, punya bawahannya hernandez, gerard dll
```

```python
run("""
SELECT 
    IFNULL(CONCAT(m.lastname, ', ', m.firstname),
            'Top Manager') AS 'Manager',
    CONCAT(e.lastname, ', ', e.firstname) AS 'Direct report'
FROM
    employees e
        LEFT JOIN
    employees m ON m.employeeNumber = e.reportsto
ORDER BY manager DESC
""")

#jika tidak ada relasi jadinya top manager
```

```python
run("""
SELECT 
    c1.city, c1.customerName, c2.customerName
FROM
    customers c1
        INNER JOIN
    customers c2 ON c1.city = c2.city
        AND c1.customername > c2.customerName
ORDER BY c1.city
""")

#customername c1 desc
```

```python
#Report the account representative for each customer.

run("""
SELECT 
    customerNumber, 
    customerName,
    firstName,
    lastName
FROM
    customers AS t1
        INNER JOIN
    employees AS t2 ON t1.SalesRepEmployeeNumber = t2.employeeNumber
""")
```

```python
#Report total payments for Atelier graphique.
run("""
SELECT 
    customerNumber, 
    customerName,
    CONCAT(m.lastname, ', ', m.firstname) AS 'Nama Sales'
FROM
    customers e
        LEFT JOIN
    employees m ON e.SalesRepEmployeeNumber = m.employeeNumber
""")
```

```python
run("""
SELECT 
    T1.customerName,
    SUM(amount) total
FROM
    customers AS T1
        INNER JOIN
    payments AS T2 ON T1.customerNumber = T2.customerNumber
    AND customerName = 'Atelier graphique'
""")
```

```python
run("""
SELECT 
    T1.customerName,
    SUM(amount) total
FROM
    customers AS T1
     JOIN
    payments AS T2 ON T1.customerNumber = T2.customerNumber
    where customerName = 'Atelier graphique'
""")
```

```python
#List products sold by order date.
run("""
SELECT 
    orderDate,
    productName,
    quantityOrdered
FROM
    orders
     JOIN orderdetails using (orderNumber)
     JOIN products Using (productCode)
ORDER BY orderDate
""")
```

```python
#What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT 
    orderNumber,
    status,
    productName,
    quantityOrdered,
    quantityInStock
FROM
    orders
     JOIN orderdetails using (orderNumber)
     JOIN products Using (productCode)
WHERE status='On Hold'
""")
```

```python
#What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT 
    productName,
    status,
    sum(quantityOrdered) as jumlahOrder,
    sum(quantityInStock) as stokTotal
FROM
    orders
     JOIN orderdetails using (orderNumber)
     JOIN products Using (productCode)
WHERE status='On Hold'
GROUP BY productName
""")
```
