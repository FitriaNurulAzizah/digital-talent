```python
import urllib

r = urllib.request.urlopen('https://lifestyle.kompas.com/').read()
print(type(r))
print(r[1:1000])
```

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(r,"lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

```python
links = soup.find_all('a')
print(len(links))

for link in links:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
berita_link = [link for link in links if '/read/' in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
import urllib
from bs4 import BeautifulSoup

html = urllib.request.urlopen("https://lifestyle.kompas.com/read/2018/10/25/110000320/4-kiat-sukses-bersaing-mendapat-pekerjaan-impian").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("h1","read__title").text.strip()
isi_berita = soup.find("div","read__content").text.strip()
tanggal = soup.find("div","read__time").text

#print(tanggal.strip())
tanggal = tanggal.strip().split('-')[1].split(',')[0].strip()
print(tanggal)
```

```python
def getBerita(link):
    html = urllib.request.urlopen(link).read()
    soup = BeautifulSoup(html, "lxml")

    judul = soup.find("h1","read__title").text.strip().encode("utf8")
    isi_berita = soup.find("div","read__content").text.strip().encode("utf8")
    tanggal = soup.find("div","read__time").text
    tanggal = tanggal.strip().split('-')[1].split(',')[0].strip()
    return [judul,isi_berita,tanggal]
```

```python
berita = []
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
    try:
        berita.append(getBerita(link["href"]))
    except:
        pass
```
