# Deskripsi Data


Dataset ini berisi mengenai pendapatan Pemerintah di APBN

Penjelasan mengenai Variabel pada Dataset ini:
* tahun: Tahun APBN
* sumber_pendapatan: Sumber Pendapatan Pemerintah (Pajak / Bukan Pajak / Hibah)
* keterangan_sumber_pendapatan: Keterangan dari Sumber Pendapatan tersebut
* nilai: Nilai dari Pos Pendapatan tersebut (dalam Milyar Rupiah)


Sumber : https://data.go.id/dataset/sumber-pendapatan-pemerintah-apbn

# Statistik tentang Data

```python
data_pendapatan.columns #Untuk mengecek header
```

```python
s = data_pendapatan.nilai.sum()
mi = data_pendapatan.nilai.min()
ma = data_pendapatan.nilai.max()
me = data_pendapatan.nilai.mean()
md = data_pendapatan.nilai.median()
mo = data_pendapatan.nilai.mode()
miss = data_pendapatan.sumber_pendapatan.min()
mass = data_pendapatan.sumber_pendapatan.max()
mist = data_pendapatan.keterangan_sumber_pendapatan.min()
mast = data_pendapatan.keterangan_sumber_pendapatan.max()

print("Total pendapatan Pemerintah di APBN\t\t:",s)
print("Rata-rata pendapatan Pemerintah di APBN\t\t:",me)
print("Median data pendapatan Pemerintah di APBN\t:",md)
print("Pendapatan tertinggi Pemerintah di APBN berasal dari",mass, "sebesar", ma, "yang bersumber dari", mast)
print("Pendapatan terendah Pemerintah di APBN berasal dari",miss, "sebesar", mi, "yang bersumber dari", mist)
```

# Visualisasi Sederhana dari Data

```python
ax = data_pendapatan.groupby('sumber_pendapatan').sum()['nilai']
ax = ax.plot(kind='pie', title ="Jumlah Pendapatan Pemerintah di APBN Berdasarkan Sumber Pendapatan ", legend=True, figsize=(20,20), fontsize=12)
```

```python
ax = data_pendapatan.groupby('tahun').sum()['nilai']
ax = ax.plot(kind='bar', title ="Jumlah Pendapatan Pemerintah di APBN per Tahun", legend=True, figsize=(20,20), fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai", fontsize=12)
```

# Narasi Analisis

Dari data di atas diperoleh informasi bahwa total pendapatan Pemerintah di bidang APBN pada tahun 1990 sampai tahun 2013 sebesar Rp 10.894.639.300.000.000 dan untuk rata-rata tiap tahunnya sebesar Rp 26.702.547.303.921,5. Jika dilihat dari grafik, pendapatan terbesar Pemerintah bersumber dari Pajak. Pendapatan yang bersumber dari bukan pajak juga cukup berperan penting, sedangkan untuk pendapatan dari hibah hanya berpengaruh sedikit. Selain itu perolehan pendapatan Pemerintah di bidang APBN dari tahun 1990 sampai tahun 2013 bisa dikatakan meningkat tiap tahunnya, dan jika dilihat secara detail pendapatan Pemerintah di bidang APBN untuk masing-masing sumber pendapatan juga meningkat tiap tahunnya. Pendapatan tertinggi Pemerintah di bidang APBN berasal dari Pajak sebesar Rp 513.509.000.000.000 yang bersumber dari Surplus Bank Indonesia. Kemudian pendapatan terendah Pemerintah di bidang APBN berasal dari Bukan Pajak yang belum mencapai Rp 1.000.000.000 yang bersumber dari Bagian Laba BUMN.
