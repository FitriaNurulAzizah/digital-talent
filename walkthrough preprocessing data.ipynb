```python
#download pip install textblob

import numpy as np
import pandas as pd
import re
from textblob import TextBlob
```

```python
split4_raw = pd.read_csv('split4.csv')
```

```python
split4_raw.head()
```

```python
split4_raw.info()
```

```python
#melihat titlenya aja
judul4_raw=split4_raw['title']
```

```python
judul4_raw
```

```python
#menghilangkan nan pada title
judul4raw=judul4_raw.dropna()
judul4raw
```

```python
article=[]
```

```python
#memasukan dataframe title pada list kosong article
for line in judul4raw:
    article.append(line.strip())
```

```python
article
```

```python
#cleansing data dan mengubah jadi huruf kecil

REPLACE_NO_SPACE = re.compile("(\.)|(\;)|(\:)|(\!)|(\'s)|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])|(\d+)|\n|\r|\ra|@|#|>>|©|(\_)|(\“)|(\-)|(\')|(\`s)|(\`)|(\â\x80\x99s)|(\â\x80\x93)|(\â\x80\x98)|(\â\x80\x99d)|(\â\x80\x99ll)|(\â\x80\x99)")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\—)|(\/)|(\“)|(\”)|(\â\x80\x99s â\x80\x98)")
NO_SPACE = ""
SPACE = " "

def preprocess_reviews(reviews):
    
    reviews = [REPLACE_NO_SPACE.sub(NO_SPACE, line.lower()) for line in reviews]
    reviews = [REPLACE_WITH_SPACE.sub(SPACE, line) for line in reviews]
    
    return reviews

article_clean = preprocess_reviews(article)
```

```python
article_clean
```

```python
#dimasukan lagi ke dataframe
news_list = pd.DataFrame.from_dict(article_clean)
```

```python
news_list
```

```python
#memberi nama kolom 'title'
news_list.columns = ['title']
```

```python
news_list
```

```python
#memberi sentiment score +1,-1,0
def analize_sentiment(title):
    analysis = TextBlob(title)
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
#menambah kolom baru SA untuk menempatkan nilai +1,-1,0
news_list['SA'] = news_list['title'].apply(analize_sentiment)
```

```python
news_list
```

```python
news_list.to_csv('split4withsa.csv',index = False)
```
