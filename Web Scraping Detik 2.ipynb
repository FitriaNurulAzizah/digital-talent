```python
import urllib

r = urllib.request.urlopen('https://news.detik.com/indeks/all').read()
print(type(r))
print(r[1:1000])
```

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(r,"lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

```python
links = soup.find_all('a')
print(len(links))

for link in links:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
berita_link = [link for link in links if '/read/' in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
import urllib
from bs4 import BeautifulSoup

html = urllib.request.urlopen("https://news.detik.com/berita/d-4276994/jembatan-suramadu-gratis-erick-thohir-jangan-hanya-komplain").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("div","jdl").find("h1").text.strip()
isi_berita = soup.find("div","detail_text").text.strip()
tanggal = soup.find("div","date").text

#print(tanggal.strip())
tanggal = tanggal.strip().split(',')[0].strip()
print(tanggal)
```

```python
gambar = soup.find("div","pic_artikel").find("img")
print(gambar['src'])

from PIL import Image
from io import BytesIO
import string

r = urllib.request.urlopen(gambar["src"]).read()
print(r)

i = Image.open(BytesIO(r))
exclude = set(string.punctuation)
judul = ''.join(ch for ch in judul if ch not in exclude)

nama_file = "images/" + judul + ".jpg"
i.save(nama_file,'JPEG')
```

```python
def getBerita(link):
    html = urllib.request.urlopen(link).read()
    soup = BeautifulSoup(html, "lxml")

    judul = soup.find("div","jdl").find("h1").text.strip().encode("utf8")
    isi_berita = soup.find("div","detail_text").text.strip().encode("utf8")
    tanggal = soup.find("div","date").text
    tanggal = tanggal.strip().split(',')[1].split('|')[0].strip()
    gambar = soup.find("div","pic_artikel").find("img")
    
    r = urllib.request.urlopen(gambar["src"]).read()
    i = Image.open(BytesIO(r))
    
    exclude = set(string.punctuation)
    judul = ''.join(str(ch) for ch in judul if ch not in exclude)
    
    nama_file = "images/" + judul + ".jpg"
    i.save(nama_file,'JPEG')
    
    return [judul,isi_berita,tanggal]
```

```python
berita = []
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
    try:
        berita.append(getBerita(link["href"]))
    except:
        pass
```
