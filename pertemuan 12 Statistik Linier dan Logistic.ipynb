```python
#Fungsi Pseudo Invers pada Statistik Linier b= (XpangkatT.X)pangkat minus 1 dikali XpangkatT dikali y (liat rumus di hp)
mones= np.ones([50,1])
print(mones)
```

```python
mones= np.ones(50)
print(mones)
```

```python
X=np.c_[X1,mones]
X.shape
```

```python
y=iris.data[0:50,1]
#50 data pertama kolom kedua
print(y)
```

```python
r1= np.matmul(X.T,y)
```

```python
r2= np.matmul(X.T,X)
```

```python
r2inv=np.linalg.inv(r2)
b= np.matmul(r2inv,r1)
print(b)
```

```python

XX=np.arange(0,10,0.02)
YY= b[0]* XX + b[1]
plt.scatter(XX,YY)
plt.show()
plt.scatter(XX, YY)
plt.scatter(iris.data[0:50,0], iris.data[0:50,1])
plt.show()
```
