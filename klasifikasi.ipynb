```python
import pandas as pd 

df = pd.read_csv('split1SA1.csv', header=0)
df
```

```python
dff=df[['title','kategori']] 
```

```python
dff
```

```python
dff.kategori.value_counts()
```

```python
# convert label to a numerical variable
dff['label_num'] = dff.kategori.map({'politik':0, 'bisnis':1})
```

```python
dff
```

```python
# split X and y into training and testing sets
from sklearn.cross_validation import train_test_split
```

```python
X = dff.title
y = dff.label_num
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# instantiate the vectorizer
vect = CountVectorizer()
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
# learn training data vocabulary, then use it to create a document-term matrix
vect.fit(X_train)
X_train_dtm = vect.transform(X_train)

# or

# equivalently: combine fit and transform into a single step
# X_train_dtm = vect.fit_transform(X_train)
```

```python
X_train_dtm
```

```python
# transform testing data (using fitted vocabulary) into a document-term matrix
X_test_dtm = vect.transform(X_test)
# X_test_dtm
```

# NAIVE BAYES

```python
# import and instantiate a Multinomial Naive Bayes model
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
# train the model using X_train_dtm (timing it with an IPython "magic command")
%time nb.fit(X_train_dtm, y_train)
```

```python
# make class predictions for X_test_dtm
y_pred_class = nb.predict(X_test_dtm)
```

```python
y_pred_class
```

```python
X_test
```

```python
y_test
```

```python
y_pred_class #[1,1] artinya kalimat 5 masuk ke kategori 1, kalimat 9 masuk ke kategori 1
```

```python
# calculate accuracy of class predictions
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
# print the confusion matrix
metrics.confusion_matrix(y_test, y_pred_class)
```

# svm

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler(with_mean = False)  
scaler.fit(X_train_dtm)

X_training_vector = scaler.transform(X_train_dtm)  
X_test_vector = scaler.transform(X_test_dtm)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
clf = svm.SVC(kernel='linear')
print("Training Classifier...")
%time clf.fit(X_training_vector, y_train)
print("Predicting...")
prediction = clf.predict(X_test_vector)
print("Prediction:",prediction)
accuracy = accuracy_score(y_test, prediction)
# metrics.accuracy_score(y_test, y_pred_class)
print('Accuracy:', accuracy)
```
